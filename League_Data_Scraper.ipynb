# Imports

```python
import urllib.request, urllib.parse, urllib.error
from datetime import datetime
import csv
import time
```

# Functions
periodic work: timer (1800s) = 30 minutes
detData: scrape important stats from op.gg

```python
def periodic_work(interval):
    while True:
        getData('https://na.op.gg/summoner/userName=HULKSMASH1337')
        getData('https://na.op.gg/summoner/userName=T1+OK+GOOD+YES')
        getData('https://na.op.gg/summoner/userName=S8+IS+SO+FUN')
        time.sleep(interval)
        
        
        
def getData(url):
    fhand = urllib.request.urlopen(url)
    text =""
    for line in fhand:
        text += line.decode()
    
    #daterecorded
    rectime = datetime.now()
    #player
    player = url.split('=')[1]                                                        #Hardcoded

    #season
    season = 10.5                                                                           #Hardcoded/unsure on s11 start
    if "2021-01-20" in str(rectime):
        season = 11

    #Ladder Position Euw
    ladderind = text.find('Ladder Rank')
    ladderpos = text[ladderind: ladderind+65: 1].rsplit("(")[1][0: 4: 1]

    #Ranked Solo
    tierind = text.find('<div class="TierRank">')
    solotier = text[tierind: tierind+35: 1].rsplit(">")[1].strip("</div")

    #LP Solo
    LPind = text.find('<span class="LeaguePoints">')
    LPscore = text[LPind: LPind+40: 1].rsplit(">")[1].strip('\n').strip('\t').strip('\n')

    #Solo wins
    soloWind = text.find('<span class="wins">')
    soloW = text[soloWind: soloWind+30: 1].rsplit(">")[1].strip("</span")

    #Solo Losses
    soloLind = text.find('<span class="losses">')
    soloL = text[soloLind: soloLind+30: 1].rsplit(">")[1].strip("</span")



    #print
    print(rectime)
    print(player)
    print(season)
    print(ladderpos)
    print(solotier)
    print(LPscore)
    print(soloW)
    print(soloL)

    with open('tylersdata.csv', 'a', newline='') as f:
        wr = csv.writer(f)
        wr.writerow([rectime,player,season,ladderpos,solotier,LPscore,soloW,soloL])

        
periodic_work(1800)
```
